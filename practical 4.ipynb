```python
import psutil

# Get all running processes with name, PID, and CPU usage
process_list = []

for proc in psutil.process_iter(['pid', 'name', 'cpu_percent']):
    try:
        process_list.append(proc.info)
    except (psutil.NoSuchProcess, psutil.AccessDenied):
        continue

# Sort processes by CPU usage (highest first)
top_processes = sorted(process_list, key=lambda x: x['cpu_percent'], reverse=True)

# Print top 5 running processes (by CPU usage)
print("Top 5 Running Processes (by CPU usage):\n")

for proc in top_processes[:5]:
    print(f"Process Name: {proc['name']}, PID: {proc['pid']}, CPU%: {proc['cpu_percent']}")
```
