```python
n = int(input("No. of processes: "))
q = int(input("Time Quantum: "))

bt = []
for i in range(n):
    bt.append(int(input(f"Burst time P{i+1}: ")))

rem = bt[:]       # remaining burst times
time = 0
ct = [0] * n      # completion times

while True:
    done = True
    for i in range(n):
        if rem[i] > 0:
            done = False
            run = min(q, rem[i])
            rem[i] -= run
            time += run
            if rem[i] == 0:
                ct[i] = time
    if done:
        break

print("\nPID\tBT\tCT\tTAT\tWT")
for i in range(n):
    tat = ct[i]
    wt = tat - bt[i]
    print(f"P{i+1}\t{bt[i]}\t{ct[i]}\t{tat}\t{wt}")
```
