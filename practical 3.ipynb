```python
import psutil

# Get virtual memory details
memory = psutil.virtual_memory()

# Convert bytes to GB
def to_gb(value):
    return value / (1024 ** 3)

print("Memory Information:\n")

print(f"Total Memory     : {to_gb(memory.total):.2f} GB")
print(f"Used Memory      : {to_gb(memory.used):.2f} GB")
print(f"Available Memory : {to_gb(memory.available):.2f} GB")
print(f"Memory Usage Percentage : {memory.percent:.2f} %")
```
