```python
import psutil
import os

print("Disk Partitions and Usage:\n")

# Get all disk partitions
partitions = psutil.disk_partitions()

for partition in partitions:
    print(f"Device: {partition.device}")
    print(f" Mountpoint: {partition.mountpoint}")
    print(f" File System Type: {partition.fstype}")

    # Skip invalid or inaccessible mountpoints
    if not os.path.exists(partition.mountpoint):
        print(" [Skipping: Invalid path]\n")
        continue

    try:
        usage = psutil.disk_usage(partition.mountpoint)
    except (PermissionError, SystemError):
        print(" [Skipping: Cannot access this partition]\n")
        continue

    print(f" Total Size: {usage.total / (1024 ** 3):.2f} GB")
    print(f" Used: {usage.used / (1024 ** 3):.2f} GB")
    print(f" Free: {usage.free / (1024 ** 3):.2f} GB")
    print(f" Percentage Used: {usage.percent}%\n")
```
