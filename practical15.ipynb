```python
requests =[55,18,98,150,38]
head = 50 
print("Head = ", head )
print(" Requests:", requests)
order = []
pending = requests.copy()
while pending:
    next_req = min(pending,key = lambda x: abs (x - head))
    order.append(next_req)
    head = next_req
    pending.remove(next_req)
print("service order :",order)
```
