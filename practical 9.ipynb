```python
pages = [1,3,0,3,5,6,3]
capacity = 3
memory = []
Faults = 0
for page in pages:
    if page not in memory:
        Faults+=1
        if len(memory) == capacity:
            memory.pop(0)
        memory.append(page)
        print(f"Page {page} -> Memory: {memory}")
        print("Total page faults(FIFO):",Faults)
```
