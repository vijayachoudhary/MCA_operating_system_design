```python
pages = [ 7, 0, 1, 2, 0, 3 ,0, 4, 2, 3, 0, 3]
capacity = 3
frames = []
page_faults = 0
for i in range(len(pages)):
    page = pages[i]
    
    if page in frames:
        print(f"{page}->hit")
        continue
    page_faults +=1
    if len(frames)<capacity:
        frames.append(page)
        print(f"{page}->page fault(added)")
    else:
        future = pages[i+1:]
        index_list = []
        for f in frames:
            if f in future:
                index_list.append(future.index(f))
            else:
                index_list.append(9999)
replace_index = index_list.index(max(index_list))
print(f"{page}-> replace {frames[replace_index]}")
frames[replace_index] = page

print("/n total page faults = ", page_faults)      
```
