```python
# Priority Scheduling (Non-Preemptive)

n = int(input("Enter number of processes: "))
processes = []

for i in range(n):
    bt = int(input(f"Burst Time of P{i+1}: "))
    pr = int(input(f"Priority of P{i+1} (lower = higher priority): "))
    processes.append([f"P{i+1}", bt, pr])

# Sort by priority (lower value = higher priority)
processes.sort(key=lambda x: x[2])

print("\nPID\tBT\tPR\tWT\tTAT")

time = 0
for pid, bt, pr in processes:
    wt = time
    tat = wt + bt
    print(f"{pid}\t{bt}\t{pr}\t{wt}\t{tat}")
    time += bt
```
