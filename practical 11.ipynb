```python
pages = [1, 2, 3, 2, 4, 1, 5, 2, 3, 4]
capacity = 3
memory = []
faults = 0

for i in range(len(pages)):
    page = pages[i]
    if page not in memory:
        faults += 1
        if len(memory) == capacity:
            next_use = {}
            for p in memory:
                if p in pages[i+1:]:
                    next_use[p] = pages[i+1:].index(p)
                else:
                    next_use[p] = float('inf')
            page_to_remove = max(next_use, key=next_use.get)
            memory.remove(page_to_remove)
        memory.append(page)
        print(f"Step {i+1}: Page {page} | Memory: {memory} (Page Fault)")
    else:
        memory.remove(page)
        memory.append(page)
        print(f"Step {i+1}: Page {page} | Memory: {memory} (Hit)")

print("Total page faults:", faults)
```
